```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'test_data\44147.ASC', header=None, skiprows=10,  
                             infer_datetime_format=True, encoding='ANSI',
                             engine='python', parse_dates=[0], skipinitialspace=True,
                             sep=None)
```

```python
df
```

```python
def cull_on_column(df, col):
    dp = np.abs(df[col]-df[col].shift())
    dp0 = np.percentile(dp.dropna(),99)
    keep = (dp>dp0) | (df.index %100 ==0)
    return keep
```

```python
keep = cull_on_column(df, 2) | cull_on_column(df, 3) | cull_on_column(df, 4)
```

```python
df[keep]
```

```python
def cull_on_column2(df, col, ncull=100):
    dp = np.abs(df[col]-df[col].shift())
    dp0 = np.percentile(dp.dropna(),99)
    idp = np.floor(df[col]/dp0)
    idp = np.abs(idp-idp.shift())
    keep = (idp==1)
    if (keep.eq(True).sum() > len(df)*6/ncull):
        print(col, "-->alternate", keep.eq(True).sum(), ">", len(df)*5/ncull)
        dp = np.abs(df[col]-df[col].shift())
        dp0 = np.percentile(dp.dropna(),99)
        keep = (dp>dp0) | (df.index % ncull ==0)
    return keep
```

```python
keep = cull_on_column2(df, 2) | cull_on_column2(df, 3) | cull_on_column2(df, 4)
```

```python
df[keep]
```

```python
def cull_on_column3(df, col, ncull=100):
    dp = np.abs(df[col]-df[col].shift())
    dp0 = np.percentile(dp.dropna(),100*(1-1/ncull))
    n=len(df)
    n0=n/ncull;
    while(n>n0):
        keep = (dp>dp0) 
        dp0 *= 1.1
        n = keep.eq(True).sum()
    return keep
```

```python
keep = cull_on_column3(df, 2) | cull_on_column3(df, 3) | cull_on_column3(df, 4)
```

```python
df[keep]
```

```python
df[keep | (df.index % 100 ==0)]
```
